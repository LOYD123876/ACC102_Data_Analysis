```python
# Import the packages we will use in this notebook.
import wrds
import pandas as pd
import numpy as np

# Create a WRDS connection.
db = wrds.Connection()

# Define the parameters we want to change easily later.
target_companies = ['MU','NVDA','AMD']
start_year = 2020
end_year = 2024

# Extract data from WRDS Compustat.
query = f"""
SELECT
    gvkey,tic AS ticker,conm AS company_name,datadate,fyear,
    ni AS net_income,
    sale AS revenue,
    ceq AS equity,
    lt AS total_liabilities,
    at AS total_assets,
    ebitda AS ebitda
FROM comp.funda
WHERE
    tic IN({','.join([f"'{t}'" for t in target_companies])})
    AND datafmt = 'STD'
    AND consol = 'C'
    AND popsrc = 'D'
    AND fyear BETWEEN {start_year} AND {end_year}
ORDER BY tic, fyear;
"""
df = db.raw_sql(query)

# Calculate the financial ratios.
df['Profit_Margin'] = df['net_income']/df['revenue']
df['Leverage_Ratio'] = df['total_liabilities']/df['equity']
df['ROE'] = df['net_income']/df['equity']
df['EBITDA'] = df['ebitda']
result_df = df[['ticker','company_name','fyear',
             'Profit_Margin','Leverage_Ratio','ROE','EBITDA']].copy()
result_df = result_df.dropna(subset=['Profit_Margin','Leverage_Ratio','ROE','EBITDA'])
display(result_df[[
    'ticker','company_name','fyear','Profit_Margin','Leverage_Ratio','ROE','EBITDA'
]])

# Save as EXCEL（optional).
result_df.to_excel("Financial_Ratios.xlsx",index=False)
db.close()
```
